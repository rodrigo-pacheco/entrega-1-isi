```python
%%sql create table medico(nif varchar(10),
                          nombre char(30),
                          especialidad char(20),
                          años_exp int(2),
                          primary key(nif));
```

```python
%%sql create table paciente(nif varchar(10),
                            nombre char(30),
                            direccion char(30),
                            edad int(2),
                            nif_medico varchar(10),
                            primary key(nif)
                            foreign key (nif_medico) references medico(nif));
```

```python
%%sql create table farmacia(nombre char(20),
                            direccion char(20),
                            telefono int(9),
                            primary key(nombre));
```

```python
%%sql create table compañia(nombre char(20),
                            telefono int(9),
                            primary key (nombre));
```

```python
%%sql create table medicamento(nombre char(20),
                               formula char(20),
                               nombre_comp char(20) not null,
                               primary key (nombre, nombre_comp),
                               foreign key (nombre_comp) references compañia(nombre) on delete cascade);
```

```python
%%sql create table vende(precio float,
                         nombre_comp char(20),
                         nombre_med char(20),
                         primary key(nombre_comp, nombre_med),
                         foreign key (nombre_comp) references compañia(nombre),
                         foreign key (nombre_med) references medicamento(nombre));
```

```python
%%sql create table receta(id int(5),
                          fecha date,
                          cantidad int,
                          primary key (id));
```

```python
%%sql create table hacer_receta(nif_pac varchar(10),
                                nif_med varchar(10),
                                id_rec int(5),
                                primary key (nif_pac, nif_med, id_rec),
                                foreign key (nif_pac) references paciente(nif),
                                foreign key (nif_med) references medico(nif),
                                foreign key (id_rec) references receta(id));
```

```python
%%sql create table receta_tiene(id_rec int(5),
                                nombre_med char(20) not null,
                                primary key (id_rec, nombre_med)
                                foreign key (id_rec) references receta(id),
                                foreign key (nombre_med) references medicamento(nombre));
```

```python
%%sql create table contrato(num_cont int(5),
                            fecha_ini date,
                            fecha_fin date,
                            condiciones string,
                            primary key (num_cont));
```

```python
%%sql create table tiene_un(nombre_farm char(20),
                            nombre_comp char(20),
                            num_cont int(5),
                            primary key (nombre_farm, nombre_comp, num_cont),
                            foreign key (nombre_farm) references farmacia(nombre),
                            foreign key (nombre_comp) references compañia(nombre),
                            foreign key (num_cont) references contrato(num_cont));
```

```python
%sql create table supervisor(nif varchar(10));
```

```python
%%sql create table asigna(nombre_farm char(20),
                          nif_sup varchar(10),
                          primary key (nombre_farm, nif_sup),
                          foreign key (nombre_farm) references farmacia(nombre),
                          foreign key (nif_sup) references supervisor(nif));
```

```python
%%sql create table supervisa(nif_sup varchar(10),
                             num_cont int(5) not null,
                             primary key (nif_sup, num_cont),
                             foreign key (nif_sup) references supervisor(nif),
                             foreign key (num_cont) references contrato(num_cont));
```
