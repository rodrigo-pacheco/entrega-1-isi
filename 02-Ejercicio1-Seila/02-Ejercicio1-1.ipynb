```python
%%sql create table enseña(dni varchar(10),
                         num_asig int,
                         num_cuatri int,
                         primary key(dni, num_asig, num_cuatri),
                         foreign key(dni) references profesores(dni),
                         foreign key(num_asig) references asignatura(num),
                         foreign key (num_cuatri) references cuatrimestre(num));
```
