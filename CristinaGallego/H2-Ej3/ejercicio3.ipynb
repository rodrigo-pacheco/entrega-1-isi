```python
%%sql 
DROP TABLE IF EXISTS departamento;
CREATE TABLE departamento(
    id INT,
    Nombre VARCHAR(30),
    Presupuesto FLOAT,
    nif_emp VARCHAR(10) NOT NULL,
    PRIMARY KEY (id),
    FOREIGN KEY(nif_emp) REFERENCES empleados(nif)
);
```

```python
%%sql 
DROP TABLE IF EXISTS empleados;
CREATE TABLE empleados(
    nif VARCHAR(10),
    salario FLOAT,
    Telefono INT,
    PRIMARY KEY(nif)
);
```

```python
%%sql 
DROP TABLE IF EXISTS tienen_hijos;
CREATE TABLE tienen_hijos(
    nombre VARCHAR(20),
    edad INT,
    nif_padre VARCHAR(10) NOT NULL,
    PRIMARY KEY (nombre, dni),
    FOREIGN KEY (nif_padre) REFERENCES empleados (nif)ON DELETE CASCADE
);
```

```python
%%sql 
DROP TABLE IF EXISTS trabajan_en;
CREATE TABLE trabajan_en(
    nif VARCHAR(10),
    id INT,
    FOREIGN KEY (nif) REFERENCES empleados(nif),
    FOREIGN KEY (id) REFERENCES departamentos(id)
);
```
