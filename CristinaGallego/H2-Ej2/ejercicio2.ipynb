```python
%%sql 
DROP TABLE IF EXISTS profesor;
CREATE TABLE profesor (
    nif VARCHAR(10),
    nombre VARCHAR(30),
    edad INT,
    cargo VARCHAR(30),
    especialidad VARCHAR(30),
    num_dep INT NOT NULL,
    PRIMARY KEY(nif),
    FOREIGN KEY (num_dep) REFERENCES departamentos(num));
```

```python
%%sql 
DROP TABLE IF EXISTS proyecto;
CREATE TABLE proyecto( 
    num INT,
    entidad VARCHAR(30),
    presupuesto FLOAT,
    nif_profesor VARCAHAR(10) NOT NULL,
    PRIMARY KEY (num),
    FOREIGN KEY (nif_profesor) REFERENCES profesor(nif));
    
```

```python
%%sql
DROP TABLE IF EXISTS estudiante;
CREATE TABLE estudiante(
    nombre VARCHAR(30),
    edad INT,
    nif VARCHAR(10),
    nif_tutor VARCHA
    R(10) NOT NULL,
    num_dep INT NOT NULL,
    PRIMARY KEY(nif),
    FOREIGN KEY (nif_tutor) REFERENCES estudiante(nif),
    FOREIGN KEY (num_dep) REFERENCES departamentos(numero));
```

```python
%%sql
DROP 
TABLE IF EXISTS departamentos;
CREATE TABLE departamentos(
    nombre VARCHAR(30),
    numero INT,
    edificio VARCHAR(30),
    PRIMARY KEY(numero),
    FOREIGN KEY (numero) REFERENCES profesor(nif));
          
```

```python
%%sql
DROP TABLE IF EXISTS trabaja;
CREATE TABLE trabaja(
    num INT,
    prof VARCHAR(30),
    PRIMARY KEY (nif_prof,num),
    FOREIGN KEY (nif_prof) REFERENCES profesor(nif),
    FOREIGN KEY (num) REFERENCES proyecto(num));
```

```python
%%sql
DROP TABLE IF EXISTS participa;
CREATE TABLE participa(
    num INT,
    nif VARCHAR(30) NOT NULL,
    PRIMARY KEY(num,nif),
    FOREIGN KEY (nif) REFERENCES estudiante(nif),
    FOREIGN KEY (num) REFERENCES proyecto(num));
```

```python

%%sql
DROP TABLE IF EXISTS supervisa;
CREATE TABLE supervisa(
    nif VARCHAR(20),
    num INT,
    nif_prof INT,
    PRIMARY KEY (nif,num, nif_prof),
    FOREIGN KEY (nif_prof) REFERENCES profesor(nif),
    FOREIGN KEY (num,nif) REFERENCES participa(num,nif));
```
