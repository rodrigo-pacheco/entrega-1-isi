Creamos las tablas para musico que tiene cuatro atributos  y esta relacionado con album, ya que un albun solo puede tener un musico.

```python
%%sql 
DROP TABLE IF EXISTS musico;
CREATE TABLE musico(
    nif  VARCHAR(10),
    nombre VARCHAR(30),
    direccion VARCHAR(30),
    telefono INT,
    id_album INT NOT NULL,
    PRIMARY KEY(nif)
);
```

Creamos la tabla de instrumento que mantiene la relacion con musico en la tabla "tiene" que creamos a continuacion  de esta

```python
%%sql 
DROP TABLE IF EXISTS instrumento;
CREATE TABLE instrumento(
    id INT,
    nombre VARCHAR(30),
    PRIMARY KEY(id)
);
```

```python
%%sql 
DROP TABLE IF EXISTS tiene;
CREATE TABLE tiene(
    nif VARCHAR(10),
    id INT,
    FOREIGN KEY (nif) REFERENCES musico(nif),
    FOREIGN KEY (id) REFERENCES instrumento(id)
);
```

Creamos la tabla para la entidad album que tiene una restriccion de participacion con musico.

```python
%%sql 
DROP TABLE IF EXISTS album;
CREATE TABLE album(
    id INT,
    titulo VARCHAR(30),
    fecha DATE,
    formato VARCHAR(30),
    nif_musico VARCHAR(10) NOT NULL,
    PRIMARY KEY (id),
    FOREIGN KEY(nif_musico) REFERENCES musicos(nif)
);
```

```python
Tenemos la relacion entre album y cancion donde cancion es una entidad debil.
```

```python
%%sql 
DROP TABLE IF EXISTS tiene_album;
CREATE TABLE tiene_album(
    titulo VARCHAR(20),
    autor VARCHAR(30),
    id_album INT NOT NULL,
    PRIMARY KEY (titulo, autor),
    FOREIGN KEY (id_album) REFERENCES album (id)ON DELETE CASCADE
);
```
